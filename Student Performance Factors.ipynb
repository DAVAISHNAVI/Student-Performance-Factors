```python
plt.figure(figsize=(15,15))

plt.suptitle("Histogram of the continous variable")

plt.subplot(2,3,1)
plt.hist(df['Hours_Studied'],edgecolor='black')
plt.title('Distribution of Hours Studied')
plt.xlabel("Hours Studied")
plt.ylabel('Frequency')

plt.subplot(2,3,2)
plt.hist(df['Attendance'],edgecolor='black')
plt.title('Distribution of Attendance')
plt.xlabel("Attendance")
plt.ylabel("Frequency")

plt.subplot(2,3,3)
plt.hist(df['Previous_Scores'],edgecolor='black')
plt.title('Distribution of Previous_Scores')
plt.xlabel('Previous_Scores')
plt.ylabel('Frequency')

plt.subplot(2,3,4)
plt.hist(df['Sleep_Hours'],edgecolor='black')
plt.title('Distribution of Sleep_Hours')
plt.xlabel('Sleep_Hours')
plt.ylabel('Frequency')

plt.subplot(2,3,5)
plt.hist(df['Exam_Score'],edgecolor='black')
plt.title('Distribution of Exam_Score')
plt.xlabel('Exam_Score')
plt.ylabel('Frequency')
plt.show()
```

# Distribution of Hours Studied:

- The distribution is approximately normal, with most students studying between 15 to 25 hours.

 - The highest frequency is around 20 hours.

- Very few students studied either less than 5 hours or more than 35 hours.

# Distribution of Attendance:

- The distribution appears to be relatively uniform, suggesting that attendance values are spread out across the range.

- There are no extreme peaks, indicating no clear preference or concentration at specific attendance percentages.

# Distribution of Previous Scores:

- The distribution is also fairly uniform, with scores ranging from 50 to 100.-
- There are slight peaks around 70 and 100, indicating some concentration of students achieving those scores.

# Distribution of Sleep Hours:

- The distribution appears to be normal (bell-shaped) with most students sleeping between 6 to 8 hours.
- The peak occurs around 7 hours, suggesting that this is the most common sleep duration.
- Few students sleep less than 5 hours or more than 9 hours, indicating that extreme sleep patterns are rare.

# Distribution of Exam Scores:

- The distribution is left-skewed, with most students scoring between 60 to 75.
- The peak frequency occurs around 65 to 70.
- Very few students have scores above 80, and scores beyond 90 are extremely rare.

# 1. Pie-chart of Physical Activity per Week

- The most common frequency of physical activity is 3 times per week (38.5%).
The second most common is 4 times per week (24.6%), followed closely by 2 times per week (23.8%).
A smaller percentage engages in physical activity 1 time per week (6.4%) or 5 times per week (5.5%).
Very few individuals engage in physical activity 6 times or 0 times per week (both less than 1%).

**Key Insight: Most individuals maintain a moderate exercise routine of 3 to 4 times per week, indicating a common pattern of balanced physical activity.**

# 2. Pie-chart of Tutoring Sessions per Month

- The most common number of tutoring sessions is 0 sessions (33.0%), indicating that a significant portion of students does not attend tutoring.
Around 25% have 2 sessions per month, and 22.9% have 1 session per month.
12.7% attend 3 sessions, while fewer students attend 4 or 5 sessions (4.6% combined).
A negligible number of students attend 6 or more sessions (close to 0%).

**Key Insight: A large portion of students does not participate in tutoring sessions, with most students opting for 0 to 2 sessions per month.**
