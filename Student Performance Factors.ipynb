```python
df.groupby(["School_Type"])["Access_to_Resources"]

sns.countplot(x=df["School_Type"],data=df,hue=df["Access_to_Resources"])
plt.title("School_Type")
plt.savefig("School_Type")
plt.show()
```

#### -The plot will show how many students with access to resources and without access belong to each School_Type. 

```python
resource_counts = df['Access_to_Resources'].value_counts()
plt.figure(figsize=(10,4))
resource_counts.plot(kind='pie', autopct='%1.1f%%',startangle=60)
plt.title('Access to Resources Distribution')
plt.show()
```

* -To show the count of Access to Resources Distribution in each level of the percentage- *

```python
sns.catplot(x='Gender',y='Exam_Score',data=df,kind="bar")
plt.show()
```

#### - The average exam score for each gender category. The height of the bars represents the mean score for each gender

```python
sns.countplot(x=df["Parental_Education_Level"],data=df,hue=df["Parental_Education_Level"])
plt.title("Parental_Education_Level")
plt.show()
```

#### -frequency of each category in the Parental_Education_Level column. Each bar will correspond to a specific education level, and the height of the bar indicates how many instances fall into that category

```python
sns.countplot(x=df["Extracurricular_Activities"],data=df,hue=df["Gender"])
plt.title("Extracurricular_Activities")
plt.show()
```

#### The plot will show the relative popularity of each extracurricular activity. identify which activities are more popular and which are less popular based on the height of the bars.

```python
sns.scatterplot(x=df['Attendance'],y=df['Exam_Score'],data=df,hue=df['Gender'])
plt.title("Attendance")
plt.show()
```

#### -How attendance correlates with exam performance (exam scores) and whether there are any gender-based patterns-
#### -Whether more attendance leads to better exam scores.-
#### -Whether gender has an impact on the relationship between attendance and exam scores.-


```python
sns.countplot(x= df['Motivation_Level'],data=df,hue=df['Motivation_Level'])
plt.title("Motivation_Level")
plt.show()
```

#### -count plot where each bar represents the count of observations for each Motivation_Level category

```python
ax=sns.countplot(x=df['Peer_Influence'],data=df,hue=df['Gender'])
ax.bar_label(ax.containers[0])
plt.title("Peer_Influence")
plt.show()
```

### count plot that visualizes how peer influence is distributed across different genders. It groups the bars by peer influence categories, separates them by gender using different colors, and adds labels to display the count of observations for each category.
